```python
#import libraries
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")
```

```python
#processing the data
hotel= pd.read_csv("C:\\Users\\sai\\OneDrive\\Documents\\Data science\\zomato.csv")
```

```python
# Removing Unwanted Columns


hotel.drop(['address','phone'],axis=1,inplace=True)
```

```python
hotel
```

```python
# Renaming and Selecting Columns in a Dataset
hotel.rename(columns = {'rate' :'rating','approx_cost(for two people)' : 'approx_cost','listed_in(type)' :'type'},inplace=True)
    
```

```python
hotel
```

```python
#Dealing with null values in a dataset
#delete null values of name column as name is the primary identifier of the dataset
# replace null values of online order with NA
# replace null values of book_table with NA
# replace null values of rating to zero as it is a numerical datatype
# replace null values of votes to zero as it is a numerical datatype
# replace null values of location to NA
# replace null values of rest_type to NA
# replace null values of dishliked to NA
# replace null values of cuisines to NA
# replace null values of approxcost to 0 as it is a numerical value
# replace null values of type to NA

hotel.isna().sum()
hotel.dropna(subset=['name'],axis=0,inplace=True)
hotel['online_order'].fillna('NA',inplace=True)
hotel['book_table'].fillna('NA',inplace=True)
hotel['rating'].fillna(0,inplace=True)
hotel['votes'].fillna(0,inplace=True)
hotel['location'].fillna('NA',inplace=True)
hotel['rest_type'].fillna('NA',inplace=True)
hotel['dish_liked'].fillna('NA',inplace=True)
hotel['cuisines'].fillna('NA',inplace=True)
hotel['approx_cost'].fillna(0,inplace=True)
hotel['type'].fillna('NA',inplace=True)
hotel.isna().sum()
```

```python
# Identifying Duplicate Data in a Dataset
# drop all the duplicate values keeping the first value as it is
hotel.duplicated().sum()
hotel.drop_duplicates(inplace=True)
```

```python
# Text Cleaning
# removing irrelevant text from all the columns


    
hotel = hotel[~hotel['name'].str.contains('RATED|Rated')]
hotel = hotel[~hotel['online_order'].astype(str).str.contains('RATED|Rated')]
hotel = hotel[~hotel['book_table'].astype(str).str.contains('RATED|Rated')]
hotel = hotel[~hotel['rating'].astype(str).str.contains('RATED|Rated')]
hotel = hotel[~hotel['votes'].astype(str).str.contains('RATED|Rated')]
hotel = hotel[~hotel['location'].astype(str).str.contains('RATED|Rated')]
hotel = hotel[~hotel['rest_type'].astype(str).str.contains('RATED|Rated')]
hotel = hotel[~hotel['dish_liked'].astype(str).str.contains('RATED|Rated')]
hotel = hotel[~hotel['cuisines'].astype(str).str.contains('RATED|Rated')]
hotel = hotel[~hotel['approx_cost'].astype(str).str.contains('RATED|Rated')]
hotel = hotel[~hotel['type'].astype(str).str.contains('RATED|Rated')]

    
```

```python
# Unique Value Check and Irrelevant Value Handling
#check for unique values in each column and handle the irrelevant values


    

hotel['online_order'] = hotel['online_order'].apply(lambda x: 'No' if x not in ['Yes', 'No'] else x)
hotel=hotel[~hotel['rating'].str.contains('[a-zS]')]
hotel['rating'] = hotel['rating'].replace(['-', 'NEW'], '0')
hotel['rating'] = hotel['rating'].str.replace('/5', '')
def convert_to_int(x):
    if x == '0':
        return 0
    else:
        return x
hotel['rating'] = hotel['rating'].apply(convert_to_int)
```

```python
# Cleaning and Exporting Zomato Dataset
# remove the unknown character from the dataset and export it to "zomatocleaned.csv"


dataframe=hotel 
dataframe['name']=dataframe['name'].str.replace('[Ãx][^A-Za-z]+','',regex=True)

#remove unknown character from dataset
    
#export cleaned Dataset to newcsv file named "zomatocleaned.csv"
dataframe.to_csv('zomatocleaned.csv', index=True)
```
